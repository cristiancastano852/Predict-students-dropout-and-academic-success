```python
#Importamos las librerias
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python

#Leemos nuestros datos
df = pd.read_csv('/content/Predict-students-dropout-and-academic-success/dataset_original.csv')
df
```

```python
#Pasamos la columna "Target" a numerica
mappings = {'Dropout': 0, 'Graduate': 1, 'Enrolled': 2}
df['Target'] = df['Target'].replace(mappings)
df
```

```python
#Eliminamos datos duplicados
df.drop_duplicates(inplace=True)
df
```

```python
#Eliminamos las filas con los valores faltantes
df.dropna(inplace=True)
df
```

```python
#Selección de características basada en la correlación
corr_matrix = df.corr().abs()
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))
to_drop = [column for column in upper.columns if any(upper[column] > 0.95)]
df.drop(to_drop, axis=1, inplace=True)
df
```

```python
correlation_matrix = df.corr()
threshold = 0.5
relevant_features = correlation_matrix[abs(correlation_matrix['Target']) > threshold]
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Matriz de correlación')
plt.show()
selected_features = relevant_features.index.tolist()
```

```python
#Vemos visualmente la columna Nacionality
df["Nacionality"].value_counts().plot(kind="bar")
```

```python
#Como la columna esta tan llena de un solo dato, se puede eliminar por ser poco util
features_tobe_removed = ["Nacionality"]

features_tobe_removed.extend(["Curricular units 1st sem (credited)", "Curricular units 1st sem (enrolled)",
       "Curricular units 1st sem (evaluations)", "Curricular units 1st sem (approved)", "Curricular units 1st sem (grade)"])
features_tobe_removed

df.drop(features_tobe_removed, axis=1, inplace=True)
df.head()
df
```

```python
# Guardamos el DataSet procesado en un archivo CSV
df.to_csv('dataset_procesado.csv', index=False)
```
