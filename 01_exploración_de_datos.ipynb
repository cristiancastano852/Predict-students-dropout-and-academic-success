## **AÃ‘ADIMOS EL DATASET**

```python
import pandas as pd

#Leemos nuestros datos
df = pd.read_csv('/content/Predict-students-dropout-and-academic-success/dataset_original.csv')
df
```

```python
df["Target"].unique()
```

```python
#Revisamos las variables que tiene nuestra columna de interes
df['Target'].value_counts()
```

```python
mappings = {'Dropout': 0, 'Graduate': 1, 'Enrolled': 2}
df['Target'] = df['Target'].replace(mappings)
df
```

```python
import matplotlib.pyplot as plt
# Crear un histograma
plt.hist(df)
# Mostrar la figura
plt.show()
```

```python
import matplotlib.pyplot as plt

counts = df.count()
plt.bar(counts.index, counts.values)
plt.xticks(rotation=90)
plt.show()
```

```python

import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize = (5,10))
sns.heatmap(df.corr()[['Target']].sort_values(by='Target', ascending=False), annot = True)
plt.title('Factors Ranker')
plt.show()
```

```python
import seaborn as sns
#Mapa de calor
sns.heatmap(df.corr())
plt.show()
```

```python
import matplotlib.pyplot as plt

# Crear el diagrama de caja y bigotes
plt.boxplot(df)

# Configurar etiquetas y tÃ­tulo
plt.xlabel('Datos')
plt.ylabel('Valores')
plt.title('Diagrama de Caja y Bigotes')

# Mostrar el diagrama de caja y bigotes
plt.show()
```
